<h1>Data Cleaning and Preparations</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
store = pd.HDFStore(os.getenv("store_data"))
```

```python
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 1000)
warnings.filterwarnings("ignore")
```

```python
data = store["airbnb_data"];data.head()
```

<h3>Removal of whitespace characters</h3>

```python
def strip_cols_and_cat_values(df: "DataFrame") -> "DataFrame":
    """ değişken adları ve kategorik değişkenlerdeki verilerden boşlukları temizleyerek yeni bir dataframe döndürür"""
    
    df.columns = df.columns.str.strip()
    
    cat_cols = df.select_dtypes(include = "object").columns
    
    df[cat_cols] = df[cat_cols].apply(lambda x: x.str.strip())
    
    return df

data = strip_cols_and_cat_values(data)
```

<h3>Filling in missing values</h3>

```python
missing_values_count = data.isnull().sum()
missing_values_count[missing_values_count != 0]
```

```python
# It is assumed that listings with room_type == Entire home/apt have at least one bathroom.

room_type1 = data[(data["room_type"] == "Entire home/apt") & (data["bathrooms"].isnull())].index
room_type2 = data[(data["room_type"] == "Private room") & (data["bathrooms"].isnull())].index
room_type3 = data[(data["room_type"] == "Shared room") & (data["bathrooms"].isnull())].index


data["bathrooms"].iloc[room_type1] = 1
data["bathrooms"].iloc[room_type2] = 0
data["bathrooms"].iloc[room_type3] = 0

data["bedrooms"] = data["bedrooms"].fillna(1)
data["beds"] = data["beds"].fillna(1)


data["host_has_profile_pic"] = data["host_has_profile_pic"].fillna("f")
data["host_identity_verified"] = data["host_identity_verified"].fillna("f")

data[["host_has_profile_pic", "host_identity_verified", "instant_bookable"]] = data[["host_has_profile_pic", "host_identity_verified", "instant_bookable"]].replace({"f": 0, "t": 1})

data["cleaning_fee"] = data["cleaning_fee"].astype("int64")

data["neighbourhood"] = data["neighbourhood"].map(lambda x: "other" if pd.isnull(x) else x)

data["host_response_rate"] = data["host_response_rate"].str.replace("%", "")
data["host_response_rate"] = data["host_response_rate"].fillna(data["host_response_rate"].astype("float").mean())
data["host_response_rate"] = data["host_response_rate"].astype("float")

data["review_scores_rating"] = data["review_scores_rating"].fillna(0)

data["city"] = data["city"].map({
                                "NYC": "New York", 
                                "SF": "San Francisco", 
                                "DC": "Washington", 
                                "LA": "Los Angeles", 
                                "Chicago": "Chicago", 
                                "Boston": "Boston"})

data = data.astype({
    "bathrooms": "int64",
    "bedrooms": "int64",
    "beds" : "int64"
})

data = data.drop_duplicates(ignore_index=True)
```

<h3>date transformation</h3>

```python
# The "last_review" and "first_review" features will be included in the model as the number of days between the first and last review dates.
data["last_review"] = pd.to_datetime(data["last_review"], format="%Y-%m-%d")
data["first_review"] = pd.to_datetime(data["first_review"], format="%Y-%m-%d")
data["first_last_review_days"] = data["last_review"] - data["first_review"]
data["first_last_review_days"] = data["first_last_review_days"].dt.days
data["first_last_review_days"] = data["first_last_review_days"].fillna(0).astype("int")

data = data.drop(["last_review", "first_review"], axis = 1)
```

<h1>Feature Selection</h1>

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.pipeline import Pipeline
from category_encoders import TargetEncoder
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV, cross_val_score, cross_validate
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from catboost import CatBoostRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
x = df.drop("log_price", axis = 1)
y = df["log_price"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=365)

for item in [x_train, x_test, y_train, y_test]: print(f"{item.shape}", end = "\n\n")
```

```python
num_cols = df.select_dtypes(include="number").drop("log_price", axis = 1).columns.to_list()

ohe_cat_cols = x_train.select_dtypes(include="object").drop(["property_type", "neighbourhood"], axis = 1).columns.to_list()
t_encode_cols = ["property_type", "neighbourhood"]


t_encoder = Pipeline([
    ("t_encoder", TargetEncoder())])

ohe_encoder = Pipeline([
    ("ohe_encoder", OneHotEncoder(sparse_output = False, handle_unknown = "ignore", drop="first"))])

scaler = Pipeline([
    ("scaler", StandardScaler())
])

cols_encode = ColumnTransformer([
    ("encode_t", t_encoder, t_encode_cols),
    ("encode_ohe", ohe_encoder, ohe_cat_cols),
    ("std_scaler", scaler, num_cols)]
    
)


cols_encode.fit(x_train, y_train)
```

```python
x_train_new = pd.DataFrame(cols_encode.transform(x_train), columns = cols_encode.get_feature_names_out())
x_test_new = pd.DataFrame(cols_encode.transform(x_test), columns = cols_encode.get_feature_names_out())
```

```python
scores_list = []
best_r2_score = -np.inf
best_k = 0
model = LinearRegression()

for k in range(1,x_train_new.shape[1] + 1):
    selector = SelectKBest(f_regression, k = k)
    selector.fit(x_train_new, y_train)
    x_train_k = selector.transform(x_train_new)

    scores = cross_val_score(estimator=model, X = x_train_k, y = y_train, cv = 10 , scoring="r2", n_jobs= -1)
    scores_list.append(scores.mean())
    if scores.mean() > best_r2_score:
        best_r2_score = scores.mean()
        best_k = k

plt.figure(figsize = (15, 7))
plt.plot(range(1, x_train_new.shape[1] + 1), scores_list)
plt.title(f"best_r2_score: {best_r2_score:.4f} | best_k: {best_k}")
plt.text(x = k, y = best_r2_score, s = ".", fontdict={"size":25});
```

```python
model_pipe = Pipeline([
    ("all_cols_transformed", cols_encode),
    ("model", LinearRegression())
])
model_pipe.fit(x_train, y_train)
```

<h1>model selection and tuning</h1>

```python
params = ([
    {
        "model": [LinearRegression()]
    },

    {
        "model": [Ridge()],
        "model__alpha": [0.01, 0.1, 1, 10, 100]
    },

    {
        "model": [KNeighborsRegressor()],
        "model__n_neighbors": [3, 5, 10, 20],
        "model__weights": ["uniform", "distance"],
        "model__metric": ["euclidean", "manhattan"]
    },

    {
        "model": [Lasso()],
        "model__alpha": [0.001, 0.01, 0.1, 1, 10]
    },

    {
        "model": [SVR()],
        "model__C": [0.1, 1, 10],
        "model__kernel": ["linear", "rbf"],
        "model__epsilon": [0.1, 0.2, 0.5]
    },

    {
        "model": [RandomForestRegressor()],
        "model__n_estimators": [200, 500, 2000],
        "model__max_depth": [10, 20, 30],
        "model__min_samples_split": [2, 5, 10]
    },
    
    
    {
        "model": [LGBMRegressor()],
        "model__n_estimators": [200, 500, 2000],
        "model__learning_rate": [0.01, 0.1, 0.3],
        "model__max_depth": [3, 5, 10]
    },

])


metrics = {
    'r2': 'r2',
    'neg_mean_squared_error': 'neg_mean_squared_error'
}


grid = GridSearchCV(
    estimator=model_pipe, 
    param_grid=params, 
    cv = 5, 
    scoring=metrics, 
    n_jobs= -1, 
    refit="r2"
)
grid.fit(x_train, y_train)
```
